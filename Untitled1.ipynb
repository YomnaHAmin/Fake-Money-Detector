```python
class Process(HasTraits):
    imgs_path_ar = Directory
    imgs_path_en = Directory
    imgs_ar = []
    imgs_en = []
    imgs = []
    
    nTotal = Int
    
    cat = Instance(Categorizor)
    det_50 = Instance(FakeDetector)
    det_100 = Instance(FakeDetector)
    det_200 = Instance(FakeDetector)
    
    load_imgs = Button
    counting = Button
    
    view = View(
        Item(
            "imgs_path_ar",  
            springy=True, 
            style='custom',
            label = "Choose Money Imgs Folder (Arabic Face)"
        ),
        Item(
            "imgs_path_en",  
            springy=True, 
            style='custom',
            label = "Choose Money Imgs Folder (English Face)"
        ),
        Item(
            "load_imgs",
            label = "Load Money Imgs"
        ),
        Item(
            "counting",
            label = "Start Counting",
            show_label = False
        ),
        Item(
            "nTotal",
            label= "Non-Fake Count of Money"
        )
    )
    
    def _imgs_path_ar_changed(self):
        for img_path in os.listdir(self.imgs_path_ar):
            if not img_path.endswith('.jpg'):
                continue
            img = cv2.imread(self.imgs_path_ar + "/" + img_path)
            self.imgs_ar.append(img)
            print img_path
        print len(self.imgs_ar)
            
    def _imgs_path_en_changed(self):
        for img_path in os.listdir(self.imgs_path_en):
            if not img_path.endswith('.jpg'):
                continue
            img = cv2.imread(self.imgs_path_en + "/" + img_path)
            self.imgs_en.append(img)
            print img_path
        print len(self.imgs_en)
            
    def _load_imgs_fired(self):
        print "Load Imgs Fired"
        print type(self.imgs_ar)
        print type(self.imgs_en)
        for i in range(0, len(self.imgs_ar)):
            self.imgs.append((self.imgs_ar[i], self.imgs_en[i]))
            print type(self.imgs_ar[i])
            print type(self.imgs_en[i])
        
    def _counting_fired(self):
        self.cat = Categorizor(self.imgs)
        categorized_imgs = self.cat.categorize()
        
        imgs_50 = categorized_imgs['50']
        imgs_100 = categorized_imgs['100']
        imgs_200 = categorized_imgs['200']
        
        self.det_50 = FakeDetector("50", imgs_50)
        self.det_100 = FakeDetector("50", imgs_100)
        self.det_200 = FakeDetector("50", imgs_200)
        
        n50 = self.det_50.detectFakeAndCount()
        n100 = self.det_100.detectFakeAndCount()
        n200 = self.det_200.detectFakeAndCount()
        
        self.nTotal = 50*n50 + 100*n100 + 200*n200
```
